```python
# leemos el archivo
df = pd.read_excel('reporte-servicios-69cb10c72ac7363b179360f3-032476cd-1f55-4dbe-aa0c-7edd7f13598c (1).xlsx')

# Visualizamos las primeras filas para entender la estructura
df.head()
```

```python
df['estado'].value_counts()
```

```python
# Convertimos las reservas a 2 tipos de estado = 'Confirmadas' y 'Cancelada'
df['estado'] = df['estado'].replace({
    'Anulada': 'Cancelada',
    'Re Confirmada': 'Confirmada',
    'Pendiente': 'Confirmada'
})
```

```python
df['estado'].value_counts()
```

```python
# Debido a fallas que ya mostro el sistema en la fase 1 de este proyecto copiamos el codigo por si volvemos a tener problemas con el estado,
# lo que hacemos aca es que si hay una fecha de cancelacion el estado pasa a ser 'Cancelado'

df.loc[df['Fecha de cancelación'].notna(), 'estado'] = 'Cancelada'
```

```python
# Nos quedamos solo con las columnas que usaremos para el proceso de automatizacion

df = df[['nombre' , 'email' , 'estado' , 'sucursal' , 'fecha', 'acepta términos y condiciones de Cluvi', 'acepta términos y condiciones del restaurante']]
```

```python
# Eliminamos reservas de personas que resevaron 2 veces para el mismo dia y cancelaron una reserva pero si visitaron el restaurante.

# Ordenar para que 'Confirmada' quede arriba
df = df.sort_values(by='estado', ascending=False)

# Eliminar duplicados reteniendo la primera coincidencia
df = df.drop_duplicates(subset=['fecha', 'email'], keep='first')
```

```python
df.info()
```

```python
# Creamos el nuevo archivo para la automatizacion con la condicional de que solo tome las reservas en estado cancelada y que aceptaron los terminos y condiciones
# para el tratamiento de datos.

df_cancelados = df[(df['estado'] == 'Cancelada') & (df['acepta términos y condiciones de Cluvi'] == 'Sí') & (df['acepta términos y condiciones del restaurante'] == 'Sí')]
```

```python
df_cancelados['fecha'] = pd.to_datetime(df_cancelados['fecha'])
```

```python
# Exportar el archivo
df_cancelados.to_excel('cancelados_automatizacion.xlsx', index=False)

# Descargarlo automáticamente al computador
from google.colab import files
files.download('cancelados_automatizacion.xlsx')
```
