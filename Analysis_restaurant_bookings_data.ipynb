```python
#Hacemos la lectura de nuestros datos que fueron limpiados y organizados previamente para dar inicio con nuestro analisis

df = pd.read_csv('datos_reservas_idilico_limpio.csv')
```

```python
# que cantidad de reservas pertenecen a cada sucursal ?

# Extraemos las fechas
inicio = df['fecha'].min()
fin = df['fecha'].max()

# Hacemos el conteo de las reservas
conteo = df['sucursal'].value_counts()
conteo_porcentajes = df['sucursal'].value_counts(normalize=True) * 100

# Guardamos los totales por su posición
total = len(df)
idilico = conteo.iloc[0]
chef = conteo.iloc[1]
porc_idilico = conteo_porcentajes.iloc[0]
porc_chef = conteo_porcentajes.iloc[1]

print(f"Desde la fecha: {inicio} hasta: {fin} Se registraron un total de: {total} reservas.\nDe las cuales {idilico} ({porc_idilico:.1f}%) pertenecen a Restaurante Idílico y {chef} ({porc_chef:.1f}%) a Mesas del Chef.")
```

```python
# Cual fue el numero de clientes que reservaron por sucursal ?

df.groupby('sucursal')['Número de personas'].sum()
personas_por_sucursal = df.groupby('sucursal')['Número de personas'].sum()
total_personas_idilico = personas_por_sucursal.iloc[1]
total_personas_chef = personas_por_sucursal.iloc[0]
total_sucursales = sum(personas_por_sucursal)

print(f"Desde la fecha: {inicio} Hasta: {fin} reservaron un total de: {total_sucursales} personas\n de las cuales {total_personas_idilico} reservaron en la sucursal Idilico para menu a la carta y\n {total_personas_chef} reservaron en la sucursal mesas del chef para menu degustación")
```
