Vaya, al parecer aqui tenemos mucho por limpiar ya que tenemos tenemos mas de 1060 datos que no tienen un estado definido, lo que haremos sera lo siguiente:

1. Vamos a manejar solo 2 estados con el fin de hacer la informacion mas facil de entender para el dueño del restaurante. (confirmada y cancelada)
2. las reservas en estado "reconfirmada" , "confirmada" y "sentado" pasaran a ser solo "confirmada"
3. las reservas en estado "el cliente no llego a la la reserva" , "anulada", "pendiente" y "cancelada" pasaran a estas en estado cancelada.
4. Aqui es donde esta el gran problema y es que casi la mitad de nuestra base de datos tiene el siguiente estado el cual parece una falla del sistema "{re_confirmed: "Re Confirmada", confirmed: "Confirmada", wait_list: "Lista de espera", pending: "Pendiente", sit: "Sentado", no_show: "El cliente no llegó a la reserva", cancelled: "Cancelada", finished: "Finalizada", annulled: "Anulada"}" y otro mensaje del sistema que es "{Translation missing: es.reservation.status_code.unknown}"
Lo que haremos sera revisar las otras columnas ( "observacion al cancelar" , "fecha de cancelacion" y "hora de cancelacion") para tratar de intuir cuales de esas reservas cancelaron y cambiarles el estado a "cancelado" y con el resto de reservas pasarlas a estado "confirmada".

*Wow, it looks like we have a lot of cleaning up to do here, as we have over 1,060 records without a defined status. Here is what we are going to do:*

*We will handle only 2 statuses in order to make the information easier to understand for the restaurant owner (confirmed and cancelled).*

*Reservations with the status "reconfirmed", "confirmed", and "seated" will become just "confirmed".*

*Reservations with the status "no-show", "annulled", "pending", and "cancelled" will be changed to "cancelled".*

*Here is where the big problem lies: almost half of our database has the following status, which looks like a system glitch: "{re_confirmed: "Re Confirmada", confirmed: "Confirmada", wait_list: "Lista de espera", pending: "Pendiente", sit: "Sentado", no_show: "El cliente no llegó a la reserva", cancelled: "Cancelada", finished: "Finalizada", annulled: "Anulada"}", alongside another system message that reads "{Translation missing: es.reservation.status_code.unknown}".*

What we will do is review the other columns ("cancellation remarks", "cancellation date", and "cancellation time") to try to infer which of those reservations were cancelled, change their status to "cancelled", and move the rest of the reservations to the "confirmed" status. *texto en cursiva* *texto en cursiva*

```python
#Antes de iniciar con la limpieza que comentamos en el paso anterior vamos a cambiar los datos que ya tenemos a solo 2 opciones ("Confirmada y Cancelada")
#Before we begin the cleanup discussed in the previous step, we will change the existing data to just 2 options ("Confirmed" and "Cancelled").
df['estado']=df['estado'].replace({'Re Confirmada': 'Confirmada' , 'Sentado' : 'Confirmada' ,'El cliente no llegó a la reserva' : 'Cancelada' , 'Anulada' : 'Cancelada' , 'Pendiente' : 'Cancelada'})
df['estado'].value_counts()
```

```python
#Vamos a tomar la columna de fecha de cancelacion y donde hayan valores osea "true" procederemos a cambiar el estado de ese dato por "Cancelada"
# We will take the cancellation date column, and where there are values (meaning "true"), we will proceed to change the status of that data to "Cancelled".
df['Fecha de cancelación'].notna()
df.loc[df['Fecha de cancelación'].notna() , 'estado' ] = 'Cancelada'
```

```python
df['estado'].value_counts()
```

```python
#Antes de repetir el paso anterior con la columna observacion al cancelar revisamos que tipo de datos tenemos en la columna
# Before repeating the previous step with the cancellation remarks column, we will review the type of data we have in the column.
df['Observación al cancelar'].value_counts()
```

```python
# Al parecer aqui tendremos que hacer una limpieza dentro de una limpieza ya que tenemos algunas reservas dobles. Vamos a proceder a eliminar
# las cancelaciones por reservas dobles, lo primero que haremos sera crear una lista con las reservas que queremos borrar
# y que nuestra columna "observaciones al cancelar" nos muestra que son reservas dobles.
# It seems we will have to do a cleanup within a cleanup here, as we have some double reservations. We will proceed to remove
# the cancellations due to double reservations; the first thing we will do is create a list of the reservations we want to delete,
# which our "cancellation remarks" column shows as double reservations.
textos_a_borrar = ['no abriremos ese dia', 'DUPLICADO', 'reserva doble', 'doble']
```

```python
# Guardamos la condición en una variable
# Save the condition in a variable
condicion_errores = df['Observación al cancelar'].isin(textos_a_borrar)

# Sobrescribimos el DataFrame quedándonos con lo que NO (~) cumple la condición
# Overwrite the DataFrame, keeping only what DOES NOT (~) meet the condition.
df = df[~condicion_errores]
```

```python
df['Observación al cancelar'].value_counts()
```

```python
#procedemos a cambiar a estado "cancelada" todas las reservas que tienen datos en la columna de 'Observación al cancelar'
# We proceed to change the status to "cancelled" for all reservations that have data in the 'Cancellation note' column

df.loc[df['Observación al cancelar'].notna() , 'estado' ] = 'Cancelada'
df['Observación al cancelar'].value_counts()
```

```python
#Luego de hacer la correcion en base a l fecha de cancelacion y obervacion al cancelar procedemos a cambiar el resto de reservas que
# muestran mensaje de error a "confirmada"
# After making the correction based on the cancellation date and cancellation note
# we proceed to change the remaining reservations showing an error message to "confirmed"

# Creamos la máscara de los que ya están bien
# We create the mask for the ones that are already correct
estados_raros = df['estado'].isin(['Confirmada', 'Cancelada'])

# Reemplazamos los que NO están bien (los errores) por 'Confirmada'
# We replace the ones that are NOT correct (the errors) with 'Confirmed'
df.loc[~estados_raros, 'estado'] = 'Confirmada'
df['estado'].value_counts()
```

```python
# Teniendo en cuenta que la columna 'Observación al cancelar' ya la usamos para cruzar los datos con 'estado' y que solo tiene 23 datos
# de mas de 3.000 reservas en total vamos a proceder a eliminarla

# Since the 'Observation upon cancellation' column has already been used to cross-reference data with 'status' and only contains 23 entries
# out of over 3,000 total reservations, we will proceed to drop it.

df = df.drop(columns=['Observación al cancelar'])
```

```python
#Continuarmos revisando el resto de columnas. Aqui enocntramos algo que tambien se debe limpiar ya que nos puede servir para una
#estrategia de mercadeo donde podamos invitar a las personas a celebrar su cumpleaños en el restaurante, pero hay demasiados datos
# asi que vamos a cambiar la informacion a 2 opciones que son (cumpleaños y aniversario).

#We will continue reviewing the remaining columns. We found another field that needs cleaning; it's valuable for a marketing strategy to
#invite guests for birthday celebrations. However, the data is too fragmented, so we will standardize it into two options:
#'Birthday' and 'Anniversary'.

df['Comentario Reserva'].value_counts()
```

```python
# Aqui convertimos las posibles opciones de personas que han ido a celebrar un cumpleaños al restaurante  en una sola que es 'Cumpleaños'
# Here, we are consolidating all the different birthday celebration entries into a single category: 'Birthday'.

condicion_cumple = df['Comentario Reserva'].str.contains('cumpleaños|cumple|birthday', case=False, na=False)
df.loc[condicion_cumple, 'Comentario Reserva'] = 'Cumpleaños'
df['Comentario Reserva'].value_counts()
```

```python
#Vamos a repetir el paso anterior pero con reservas que van a celebrar "Aniversario" o "Anniversary"
# Repeating the previous step for reservations celebrating an 'Anniversary'.

condicion_aniversario = df['Comentario Reserva'].str.contains('aniversario|anniversary|aniversary|aniversa', case=False, na=False)
df.loc[condicion_aniversario, 'Comentario Reserva'] = 'Aniversario'
df['Comentario Reserva'].value_counts()
```

```python
# Luego de tener los datos de "Cumpleaños" y "Aniversario" que nos puede servir para estrategias de mercadeo vamos a eliminar
# el resto de comentarios de la columna 'Comentario Reserva'
# After isolating the 'Birthday' and 'Anniversary' data for marketing strategies, we will drop the remaining entries from the
#'Reservation Comment' column.

eventos_clave = df['Comentario Reserva'].isin(['Cumpleaños', 'Aniversario'])
df.loc[~eventos_clave, 'Comentario Reserva'] = None
df['Comentario Reserva'].value_counts()
```
