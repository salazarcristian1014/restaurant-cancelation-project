```python
# Guardamos la ruta del archivo en una variable
# We authorize Colab to access our Google Drive files.
ruta_archivo = '/content/drive/MyDrive/Reservas_Restaurante/Reservas_Restaurante.csv'

# Leemos el archivo CSV y lo convertimos en nuestro DataFrame llamado 'df'
# We read the CSV file and convert it into our DataFrame named 'df'
df = pd.read_csv(ruta_archivo)

# La función .head() nos muestra las primeras 5 filas para verificar que todo cargó bien
# The .head() function shows us the first 5 rows to verify that everything loaded correctly
df.head()
```

Procedemos a eliminar las columnas que no tienen valores en ninguna de sus filas y luego revisamos que si se hayan eliminado.

*We proceed to drop the columns that have no values in any of their rows, and then we verify that they have been removed.*

```python
df=df.dropna(axis = 1, how='all')
df.info()
```

Eliminamos desde la columna 26 hasta la columna 38 ya que estas columnas nos aportan datos de facturacion electronica y permisos legales que solicita el sistema de reservas y esta informacion no es util para el problema que queremos resolver

*We drop columns 26 through 38, as these columns provide electronic billing data and legal permissions requested by the reservation system, and this information is not useful for the problem we want to solve.*

```python
#Revisamos los nombres de las columnas para copiarlos y pegarlos de una manera mas practica en la proxima linea de codigo
# We check the column names so we can copy and paste them more practically in the next line of code
df.columns[26:39]
```

```python
#Eliminamos las columnas con informacion legal y datos de facturacion electronica
# We drop the columns containing legal information and electronic billing data
df = df.drop(columns =['FE documento', 'FE Nombre', 'FE email', 'FE teléfono', 'FE billingId',
       'acepta términos y condiciones de Cluvi',
       'fecha de aceptación de términos y condiciones de Cluvi',
       'hora de aceptación de términos y condiciones de Cluvi',
       'términos y condiciones de Cluvi',
       'acepta términos y condiciones del restaurante',
       'fecha de aceptación de términos y condiciones del restaurante',
       'hora de aceptación de términos y condiciones del restaurante',
       'términos y condiciones del restaurante'] )
```

```python
# Revisamos nuevamente como va la limpieza de nuestro archivo
# We check again how our file cleaning is progressing
df.info()
```

Eliminamos las columnas de Alergia, responsable, llego y observacion ya que tienen muy pocos datos y ademas no estan relacionados con temas de cancelaciones y revisamos nuevamente como va nuestro archivo.
*We drop the 'Alergia' (Allergy), 'responsable' (Responsible), 'llego' (Arrived), and 'observacion' (Observation) columns since they have very little data and are not related to cancellation issues, and we check our file again.*

```python
df = df.drop(columns = ['Alergias', 'Responsable', 'llego', 'observación'])
df.info()
```

Aqui teniendo en cuenta que ya sacamos al informacion basura que no tenia datos, columnas que no nos servian para el analisis y columnas con muy pocos datos... Implementamos algo de logica para el logro principal de este proyecto que es conocer cual es la tasa de cancelacion en base al total de reservas por esta razon vamos a revisar los datos de cada una de las columnas y miramos si la dejamos o la descartamos.

*Here, considering that we have already removed the junk information that had no data, columns that were useless for the analysis, and columns with very little data... We implement some logic for the main goal of this project, which is to determine the cancellation rate based on total reservations. For this reason, we will review the data in each of the columns and see if we keep it or discard it.*

```python
#El codigo lo dejamos ya que nos sirve como contador de numero total de reservas
# We keep the 'codigo' (code) column as it serves as a counter for the total number of reservations
df['codigo'].value_counts()
```

```python
#La sucursal la dejamos ya que nos puede ayudar a segmentar los datos y conocer en que sucursal se presentan mayor numero de cancelaciones
# We keep the 'sucursal' (branch) column as it can help us segment the data and find out which branch has the highest number of cancellations
df['sucursal'].value_counts()
```

```python
#Con este codigo podemos ver desde y hasta que fecha tenemos los datos, por esa razon vamos a organizar
# el formato para que no se lea como object sino como datetime
# With this code, we can see the date range of our data; for that reason, we are going to format
# it so it is not read as an 'object' but as 'datetime'
df['fecha'].value_counts()
```

```python
# Convertimos la columna fecha a tipo datetime
# We convert the 'fecha' (date) column to datetime type
df['fecha'] = pd.to_datetime(df['fecha'])
```

```python
# La columna hora la dejaremos ya que nos puede dar datos de cuales son los horarios mas congestionados y al igual que con la fecha
# la convertimos de dato object a datetime
# We will keep the 'hora' (time) column as it can provide data on the busiest hours, and just like with the date,
# we convert it from 'object' data type to 'datetime'
df['hora'] = pd.to_datetime(df['hora'])
df['hora'].value_counts()
```

```python
# Esta columna es vital para calcular el numero de personas que Reservaron ya que pueden haber muchas reservas registradas pero
# solo conociendo el numero de personas por reserva podremos saber en total cuantas personas reservaron.
# This column is vital for calculating the number of people who booked, since there might be many registered reservations, but
# only by knowing the number of people per reservation will we be able to know the total number of people who booked.
df['Número de personas'].value_counts()
```

```python
#Esta variable nos puede ayudar a entender si el porcentaje de cancelaciones es mayor en nuevas personas o en personas que ya nos han
# This variable can help us understand if the cancellation percentage is higher among new guests or returning customers
df['Nuevo usuario'].value_counts()
```

```python
df.info()
```

Las columnas con nombre, email, codigo de telefono y telefono a pesar de que no las necesitamos para el analisis inicial sobre la tasa de cancelacion son de vital importancia para la segunda parte del proyecto donde buscaremos al forma de automatizar esas reservas canceladas e invitarlas a que nos digan el motivo de la cancelacion y ademas poder reinvitarlos al restaurante.

*The columns containing name, email, country code, and phone number, although not needed for the initial cancellation rate analysis, are of vital importance for the second part of the project. There, we will look for a way to automate outreach to those canceled reservations, inviting them to tell us the reason for the cancellation, and also being able to invite them back to the restaurant.*

```python
# Procedemos a modificar el tipo de dato int de la columna 'telefono' para luego poder concatenarlo con 'codigo' y al momento
# de la automatizacion se puedan enviar mensajes por whatsapp
# We proceed to modify the 'int' data type of the 'telefono' (phone) column so we can later concatenate it with 'codigo' (country code), so that at the time
# of the automation, we can send messages via WhatsApp
df['teléfono'] = df['teléfono'].astype(str)
```

```python
# Creamos la columna 'telefono_completo' esto con el fin de concatenar el codigo y el telefono para poder enviar mensajes via whatsapp
# We create the 'telefono_completo' (full phone) column in order to concatenate the country code and the phone number to be able to send messages via WhatsApp
df['telefono_completo'] = '+' + df['código de teléfono'] + df['teléfono']
```

```python
df['telefono_completo'].value_counts()
```

```python
# Procedemos a eliminar la columna 'telefono' ya que al tener el 'telefono_completo' no es necesaria tenerla en nuestro df, dejamos la
# columna de codigo ya que nos puede servir para identificar de que paises nos visitan en su mayoria.
# We proceed to drop the 'telefono' column since having 'telefono_completo' makes it unnecessary in our df. We keep the
# 'codigo' column as it can help us identify which countries the majority of our visitors come from.
df = df.drop(['teléfono'], axis = 1)
df.info()
```

Teniendo en cuenta que la columna que creamos telefono_completo quedo al final y lo ideal es que quede junto con la informacion personal vamos a organizar de tal forma que se pueda mover y quede antes de 'estado'

*Considering that the 'telefono_completo' column we created ended up at the end, and ideally it should be alongside the personal information, we are going to reorganize it so it can be moved right before 'estado' (status).*

```python
#revisamos las columnas que tenemos actualmente
# We check the columns we currently have
df.columns
```

```python
# teniendo en cuenta que la columna que creamod con 'telefono_completo' quedo al final y con el fin de hacerlo de una forma mas organizada
# procedemos a reestructurar el orden de las columnas de tal forma que la informacion personal quede lo mas junta posible.
# Considering that the column we created with 'telefono_completo' ended up at the end, and in order to make it more organized,
# we proceed to restructure the column order so that the personal information is grouped as closely together as possible.
nuevo_orden = [
    'codigo', 'sucursal', 'fecha', 'hora', 'Número de personas', 'Nuevo usuario',
    'nombre', 'email', 'código de teléfono', 'telefono_completo', 'estado',
    'Comentario Reserva', 'Observación al cancelar', 'Tipo de reserva', 'origen', 'salió',
    'Fecha de creación', 'Hora de creación', 'Fecha de cancelación', 'Hora de cancelación',
    'Cobros', 'Pagado'
]

# Aplicamos el nuevo orden al DataFrame
# We apply the new order to the DataFrame
df = df[nuevo_orden]

# Verificamos como va nuestro archivo
# We verify how is our file dateframe going
df.info()
```
