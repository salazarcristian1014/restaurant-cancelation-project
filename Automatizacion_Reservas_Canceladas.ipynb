```python
import pandas as pd
df_correos = pd.read_excel('cancelados_automatizacion.xlsx')
```

```python
from email.message import EmailMessage
```

```python
# Lo primero sera conectarnos con el correo coorporativo para que los email se vean mas profesionales.

import smtplib
remitente = "contacto@restaurante.com"
contrasena = "SECRETO"
```

```python
# Identificamos la ruta que seguiremos para la conexion con nuestro correo empresarial.
servidor_smtp = "smtp.hostinger.com"
puerto = 465
```

```python
#Conectamos con el servidor donde tenemos alojado nuestro correo electronico empresarial.
conexion = smtplib.SMTP_SSL(servidor_smtp, puerto)
conexion.login(remitente,contrasena)
```

```python
# Luego creamos una condicion for para recorrer cada una de las lineas de nuestro archivo ya limpio
# y poder enviar un correo electronico que nos ayude a identificar el motivo principal de cancelaciones

for indice, fila in df_correos.iterrows():
 nombre = fila['nombre']
 email = fila['email']
 mensaje = f"Hola {nombre}, somos Cristian y Yeison fundadores de Restaurante Idilico, notamos que Cancelaste la reserva que tenias con nosotros la semana pasada, como apasionados por la gastronomia pero tambien emprendedores el feedback de nuestros clientes es muy importante, nos apoyarias diligenciando la encuesta adjunta seguro no te toma mas de 1 minuto. Lo mas probable es que te surgiera algo de ultimo momento por eso te dejamos el link para que puedas hacer tu reserva y darnos el gusto de atenderte en una proxima ocasion."
 sobre = EmailMessage()
 sobre['Subject'] = f"{nombre}, nos hizo falta tu visita esta semana en Idílico 🍷"
 sobre['From'] = remitente
 sobre['To'] = email
 sobre.set_content(mensaje)
```
